```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies = movies[['id','title','overview','popularity','genres','keywords','vote_average','vote_count']]
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['keywords']
movies['tags'] = np.array(movies['tags'])
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies_new = movies.drop(columns=['id','overview','genres','keywords'])
```

```python
movies_new.head()
```

# Splitting Training

```python
med=movies_new['vote_count'].median()
med
```

```python
x=movies_new.loc[:,['popularity','vote_average']]
x.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
inputarray = cv.fit_transform(movies_new['tags']).toarray()
inputarray
```

```python
inputarray.shape
```

```python
from sklearn.preprocessing import normalize
x = normalize(x)
x=pd.DataFrame(x,columns=['popularity','vote_average'])
x.head()
```

```python
y=movies_new.loc[:,['vote_count']]
y['vote_count'] = (y['vote_count'] >= med).astype(int)
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=5)
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs = cosine_similarity(inputarray)
cs
```

```python
def recommend(movie):
    index = movies_new[movies_new['title'] == movie].index[0]
    distances = sorted(list(enumerate(cs[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

# K- Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
#%matplotlib inline
```

```python
movies.head()
```

```python
movies2=movies.drop(columns=['id','title','overview','genres','keywords','tags'])
movies2=normalize(movies2)
movies2=pd.DataFrame(movies2,columns=['popularity','vote_count','vote_average'])
movies2.head()
```

```python
plt.scatter(movies2.popularity,movies2['vote_count'])
plt.xlabel('popularity')
plt.ylabel('vote_count')
```

```python
km =KMeans(n_clusters=3)
y_pred=km.fit_predict(movies2[['popularity','vote_count']])
y_pred
```

```python
movies2['cluster']=y_pred
movies2.head()
```

```python
km.cluster_centers_
```

```python
df1 = movies2[movies2.cluster==0]
df2 = movies2[movies2.cluster==1]
df3 = movies2[movies2.cluster==2]
plt.scatter(df1.popularity,df1['vote_count'],color='green')
plt.scatter(df2.popularity,df2['vote_count'],color='red')
plt.scatter(df3.popularity,df3['vote_count'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('popularity')
plt.ylabel('vote_count')
plt.legend()
```
